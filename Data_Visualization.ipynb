# **Data Visualization**

<h3><b>Tugas 1</b></h3>
<h4><b>Visualisasi</b></h4>
<h4>Distribusi Usia (AGE) Pelanggan</h4>

```python
print('Bentuk Visualisasi Plot \033[1m\033[96mDensity\033[96m\033[1m\n')

# Plot Options
sns.set_style('white')
plt.figure(figsize=(10,5))

# Buat Plot
sns.kdeplot(df_1['AGE'], color = "orange", fill = True)

# Buat Garis Vertikal sebagai Penanda letak Nilai Median dari Distribusi Usia Pelanggan
plt.axvline(df_1['AGE'].median(), ymax = 0.95, linestyle = "--")

# Masukkan Nilai Median dari Distribusi Usia Pelanggan
plt.text(df_1['AGE'].median() + 2, 0.025, 'Median: ' + str(df_1['AGE'].median()), fontsize = 10, style = "italic", fontweight = "bold")

# Tampilkan Plot
plt.yticks([])
plt.xlabel('Usia Pelanggan', fontsize = 12, fontweight = "semibold")
plt.ylabel('Density')
plt.title('DISTRIBUSI USIA PELANGGAN', fontsize = 14, fontweight = "semibold")
plt.show()

print('\n')

import warnings
warnings.filterwarnings("ignore")

print('Bentuk Visualisasi Plot \033[1m\033[96mDensity\033[96m\033[1m dan \033[1m\033[96mHistogram\033[96m\033[1m\n')

sns.distplot(df_1.AGE,kde=True,color='red',
             hist_kws={'edgecolor':'white','linewidth':1},
             kde_kws={'color':'black', 'linewidth':2})
plt.title('DISTRIBUSI USIA PELANGGAN', fontsize = 14, fontweight = "semibold")
plt.xlabel('Usia Pelanggan', fontsize = 12, fontweight = "semibold")
plt.show()
```

```python
# Group the data by category and calculate the sum of calories for each category
grouped_df = df_2.groupby(['Category']).sum()

# Calculate the percentage of calories contributed by beverages (soda and coffee)
beverage_calories = grouped_df.loc['Beverages', 'Calories']
total_calories = grouped_df['Calories'].sum()
beverage_calorie_pct = round((beverage_calories / total_calories) * 100, 2)

# Print the results
print('Beverages (soda and coffee) contribute', beverage_calorie_pct, '% of the total calories in McDonald\'s food items.')
```

```python
# Create a pie chart to visualize the percentage of total calories from beverages
labels = ['Beverages', 'Other Foods']
sizes = [beverage_calorie_pct, 100 - beverage_calorie_pct]
colors = ['lightskyblue', 'lightcoral']
plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=90)
plt.axis('equal')
plt.title('Percentage of Total Calories from Beverages in McDonald\'s Food Items')
plt.show()
```

```python
# Create a bar chart to visualize the contribution of beverages to the overall caloric intake
labels = ['Beverages', 'Other Foods']
sizes = [beverage_calorie_pct.sum(), 100 - beverage_calorie_pct.sum()]
colors = ['lightskyblue', 'lightcoral']
plt.bar(labels, sizes, color=colors)
plt.title('Contribution of Beverages to the Overall Caloric Intake in McDonald\'s Nutrition Facts Dataset')
plt.ylabel('Percentage of Total Calories')
plt.show()
```

<h3><b>Tugas 3</b></h3>
<h4><b>Visualisasi</b></h4>
<h4>Perbandingan jumlah nutrisi yang terkandung dalam suatu produk makanan sandwich dengan isian antara ayam panggang (<b>Grilled</b>) dan isian ayam renyah (<b>Crispy</b>)</h4>

```python
# Does ordered grilled chicken instead of crispy increase a sandwich's nutritional value?
# Create data subsets for grilled and crispy chicken
df_grilled = df_2[df_2['Item'].str.contains('Grilled Chicken') & df_2['Item'].str.contains('Sandwich')]
df_crispy = df_2[df_2['Item'].str.contains('Crispy Chicken') & df_2['Item'].str.contains('Sandwich')]

# Calculating the average nutritional value for grilled and crispy chicken
mean_grilled = df_grilled.mean()
mean_crispy = df_crispy.mean()

# Calculating nutrition
nutrient_score_grilled = mean_grilled['Protein'] - mean_grilled['Total Fat'] + mean_grilled['Dietary Fiber']
nutrient_score_crispy = mean_crispy['Protein'] - mean_crispy['Total Fat'] + mean_crispy['Dietary Fiber']

# Comparing nutrition
if nutrient_score_grilled > nutrient_score_crispy:
  print("Yes, ordered grilled chicken instead of crispy increase a sandwich's nutritional value")
elif nutrient_score_grilled < nutrient_score_crispy:
  print("No, ordered grilled chicken instead of crispy doesn't increase a sandwich's nutritional value")
else:
  print("Grilled Chicken Sandwich is as nutritious as Crispy Chicken Sandwich")
```

```python
# Filter the dataset to only include sandwiches with chicken
chicken_sandwiches = df_2[df_2['Item'].str.contains('Chicken') & df_2['Item'].str.contains('Sandwich')]

# Filter the dataset to only include crispy and grilled chicken sandwiches
crispy_chicken_sandwiches = chicken_sandwiches[chicken_sandwiches['Item'].str.contains('Crispy Chicken')]
grilled_chicken_sandwiches = chicken_sandwiches[chicken_sandwiches['Item'].str.contains('Grilled Chicken')]

# Create a new DataFrame to hold the nutritional values
data = pd.DataFrame({
    'Calories': {
        'Crispy Chicken': crispy_chicken_sandwiches['Calories'].mean(),
        'Grilled Chicken': grilled_chicken_sandwiches['Calories'].mean()
    }
})

# Reshape the data for plotting
data = data.reset_index().melt(id_vars='index', var_name='Nutrition')

# Create a bar plot to compare the nutritional value of crispy and grilled chicken sandwiches
sns.set(style='whitegrid')
g = sns.catplot(x='index', y='value', hue='Nutrition', data=data, kind='bar', palette='muted')
g.despine(left=True)
g.set_xlabels('Sandwich Type')
g.set_ylabels('Calories')
g.set(title='Comparison of Calories for Crispy and Grilled Chicken Sandwiches')

plt.show()
```

```python
# Exploring the amount of calories on Crispy Chicken category
crispy = df_2[df_2['Item'].str.contains('Crispy Chicken')]
crispy_cal = pd.DataFrame({'Item': crispy.Item, 'Calories': crispy.Calories})
crispy_sort = crispy_cal.sort_values('Calories', ascending=False)
ax = crispy_sort.plot.barh(x='Item', y='Calories', figsize= (10,7))
ax.set_title("Total Calories in Crispy Chicken Item Category (in Cal)")
```
